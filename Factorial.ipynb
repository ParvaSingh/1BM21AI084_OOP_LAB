```python
import math

class Number:
    def __init__(self, value):
        self.value = float(value)

    def isZero(self):
        return self.value == 0

    def isPositive(self):
        return self.value > 0

    def isNegative(self):
        return self.value < 0

    def isOdd(self):
        return self.value % 2 != 0

    def isEven(self):
        return self.value % 2 == 0

    def isPrime(self):
        if self.value < 2:
            return False
        for i in range(2, int(math.sqrt(self.value)) + 1):
            if self.value % i == 0:
                return False
        return True

    def isArmstrong(self):
        num_str = str(int(self.value))
        num_digits = len(num_str)
        total = sum(int(digit) ** num_digits for digit in num_str)
        return total == self.value

    def getFactorial(self):
        if self.value < 0:
            return "Undefined (Factorial is not defined for negative numbers)"
        result = 1
        for i in range(1, int(self.value) + 1):
            result *= i
        return result

    def getSqrt(self):
        return math.sqrt(self.value)

    def getSquared(self):
        return self.value ** 2

    def sumDigits(self):
        return sum(int(digit) for digit in str(int(self.value)))

    def getReverse(self):
        return float(str(int(self.value))[::-1])


# Driver class
def main():
    num1 = Number(5)
    num2 = Number(-8)
    num3 = Number(13)

    # Test methods
    print(f"Is Zero? {num1.isZero()}")
    print(f"Is Positive? {num1.isPositive()}")
    print(f"Is Negative? {num1.isNegative()}")
    print(f"Is Odd? {num1.isOdd()}")
    print(f"Is Even? {num1.isEven()}")
    print(f"Is Prime? {num1.isPrime()}")
    print(f"Is Armstrong? {num1.isArmstrong()}")

    print(f"\nFactorial: {num2.getFactorial()}")
    print(f"Sqrt: {num2.getSqrt()}")
    print(f"Squared: {num2.getSquared()}")
    print(f"Sum of Digits: {num2.sumDigits()}")
    print(f"Reverse: {num2.getReverse()}")


if __name__ == "__main__":
    main()

```
